```python
import numpy as np 
X = np.arange(4)
Y1 = X * X
Y2 = 1.2 * X * X
Y3 = 0.8 * X * X
plt.plot(X, Y1)
plt.plot(X, Y2, color='m', marker='o', linestyle='--')
plt.plot(X, Y3, 'k^:')
plt.grid()
plt.show()
```

```python
plt.plot(X, Y1, label='Y1')
plt.plot(X, Y2, color='m', marker='o', linestyle='--', label='Y2')
plt.plot(X, Y3, 'k^:', label='Y3')
plt.grid()
plt.legend()
plt.show()
```

## 스타일

스타일 문자열    약자    의미
color              c   선 색깔
linewidth         lw   선 굵기
linestyle         ls  선 스타일
marker      마커 종류  markersize
ms          마커 크기  markeredgecolor
mec        마커 선 색깔 markeredgewidth
mew        마커 선 굵기 markerfacecolor
mfc     마커 내부 색깔 markerfacecolor

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("x축, y축 범위지정")
plt.xlim(0, 50)
plt.ylim(-5, 20)
plt.show()
```

## 틱 설정

```python
import numpy as np 
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```
